```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import random
import time
```

```python
def web_crawling(keyword):
    google_page = driver.get('https://www.google.com/')                  #Opens a google page in chrome browser
    search_box = driver.find_element_by_xpath('//input[@name="q"]')     # Find the google search box
    search_box.send_keys(keyword)                                      #Enter the keyword  "Python"
    search_box.send_keys(Keys.ENTER)   
    #x = input('输入')
    pages_links =[]

    while True:
        
        next_page_btn = driver.find_elements_by_id("pnnext")
        if len(next_page_btn) < 1:
            print("No more pages left")
            break
        
        time.sleep(random.uniform(10,20))
        next_page_btn[0].click()                                                                 #Click that page to go that page
    #     except NoSuchElementException:
    #         driver.quit()   
    
        value = driver.find_elements_by_xpath("//div[@class='g']//div[@class='yuRUbf']//a[not(@class)]")    #google mainpage 
        #value = driver.find_elements_by_xpath("//h3[@class='gs_rt']/a")                                      #google scholar

        #Gets Links of that page
        print(value)
        ## Loop for getting the href value of all the links in search result
        for each_val in value:
            link = each_val.get_attribute('href')
            pages_links.append(link)
    
    filename = keyword + '.txt'
    with open(filename, 'w') as f:
        f.write('\n'.join(pages_links))
```
